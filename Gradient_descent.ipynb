```python
import numpy as np
```

```python
#to find the best fit line between the points of x and y
def grad(x,y):
  #satrt with some values of m_curr and b_curr
  m_curr = b_curr = 0
  #define how many steps we are going to do
  iterations = 1000
  n = len(x)
  #define the learning rate
  learning_rate = 0.001
  #loop through iterations(each step) and calculate the y_predicted
  for i in range(iterations):
    y_predicted = m_curr * x + b_curr #y = mx + b
    cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
    md = -(2/n) * sum(x * (y - y_predicted))
    bd = -(2/n) * sum(y - y_predicted)
    m_curr = m_curr - learning_rate * md
    b_curr = b_curr - learning_rate * bd
    print("m {}, b {}, iteration {}".format(m_curr,b_curr,cost,i))




x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
```
