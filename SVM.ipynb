```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
#creating a df from the above data
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
#append target vairalbe to the dasta set since its acialable on desc
df["target"] = iris.target
df.head()
```

```python
iris.target_names #[0, 1,2]
```

```python
#to see how many rows have value 1
df[df.target==1].head()
```

```python
df[df.target==0].head()
```

```python
df[df.target==2].head()
```

```python
# adding flower column for a clarification
df["flower_name"] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
# importing and training the model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(["target", "flower_name"], axis="columns"), df.target, test_size=0.2)
```
