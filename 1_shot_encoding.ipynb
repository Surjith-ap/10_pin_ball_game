```python
import pandas as pd
df =  pd.read_csv("housing_data.csv")
df
```

```python
#creating dummies
dummies = pd.get_dummies(df.town)
```

```python
#concatinating dummies into main
merged = pd.concat([df,dummies],axis='columns')
merged
```

```python
# dropping original town column and 1=one of the newly created columns to
# avoid dimmies problem
final = merged.drop(['town','west windsor','id'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# x is the input given
x = final.drop('price',axis='columns')
x
```

```python
#and alos y
y = final.price
y
```

```python
# training the model
model.fit(x,y)
```

```python
model.predict([[2800,1,0]])
```

```python
model.predict([[2800,0,1
                ]])
```

```python
# to check the accuracy of the model
model.score(x,y)
```
